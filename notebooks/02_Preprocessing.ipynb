```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Settings
pd.set_option('display.max_columns', None)
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# Load cleaned data from Day 1
df = pd.read_csv('../data/train_cleaned.csv')

print("Libraries imported!")
print(f"Dataset shape: {df.shape}")
df.head()
```

```python
# Check missing values
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
missing_pct = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing Percent': missing_pct
})

print("Columns with Missing Values:")
print(missing_df)
print(f"\nTotal columns with missing values: {len(missing_df)}")
```

```python
# Fix numerical missing values with median
num_cols = df.select_dtypes(include=np.number).columns.tolist()
num_cols.remove('SalePrice')

for col in num_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)
        
print("Numerical missing values fixed!")
print(f"Remaining missing in numerical cols: {df[num_cols].isnull().sum().sum()}")
```

```python
# Fix categorical missing values with mode (updated for new pandas)
cat_cols = df.select_dtypes(include='str').columns.tolist()

# If no str cols found, use object
if len(cat_cols) == 0:
    cat_cols = [col for col in df.columns if df[col].dtype == 'object']

for col in cat_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])

print("Categorical missing values fixed!")
print(f"Remaining missing in categorical cols: {df[cat_cols].isnull().sum().sum()}")
print(f"\nTotal missing values in dataset: {df.isnull().sum().sum()}")
```

```python
# Check which numerical columns still have missing values
still_missing = df.isnull().sum()
still_missing = still_missing[still_missing > 0].sort_values(ascending=False)
print("Columns still having missing values:")
print(still_missing)
```

```python
# Fix remaining numerical missing values
for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype in ['float64', 'int64']:
            df[col] = df[col].fillna(df[col].median())

print("All missing values fixed!")
print(f"Total missing values remaining: {df.isnull().sum().sum()}")
```

```python
# Create new features
df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']
df['HouseAge'] = df['YrSold'] - df['YearBuilt']
df['RemodAge'] = df['YrSold'] - df['YearRemodAdd']
df['TotalBath'] = (df['FullBath'] + df['HalfBath'] * 0.5 +
                   df['BsmtFullBath'] + df['BsmtHalfBath'] * 0.5)
df['HasGarage'] = (df['GarageArea'] > 0).astype(int)
df['HasBasement'] = (df['TotalBsmtSF'] > 0).astype(int)
df['HasPool'] = (df['PoolArea'] > 0).astype(int)

print("New features created!")
print(f"Dataset shape after feature engineering: {df.shape}")
print("\nNew features added:")
print("TotalSF - Total square footage")
print("HouseAge - Age of house")
print("RemodAge - Years since remodel")
print("TotalBath - Total bathrooms")
print("HasGarage - Whether house has garage")
print("HasBasement - Whether house has basement")
print("HasPool - Whether house has pool")
```

```python
from sklearn.preprocessing import LabelEncoder

# Get categorical columns
cat_cols = [col for col in df.columns if df[col].dtype == 'object']
print(f"Total categorical columns to encode: {len(cat_cols)}")

# Apply Label Encoding
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col].astype(str))

print("All categorical columns encoded!")
print(f"Dataset shape: {df.shape}")
print(f"Total missing values: {df.isnull().sum().sum()}")
```

```python
# Check all datatypes
print("Data types in dataset:")
print(df.dtypes.value_counts())
print(f"\nTotal columns: {df.shape[1]}")
print(f"Total rows: {df.shape[0]}")
print(f"\nAll columns are numerical - Ready for modeling!")
```

```python
# Find columns that are still strings
string_cols = [col for col in df.columns if df[col].dtype == 'object' 
               or df[col].dtype.name == 'string']
print(f"String columns found: {len(string_cols)}")
print(string_cols)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Fix all string/object columns
for col in df.columns:
    if df[col].dtype == 'object' or df[col].dtype.name == 'string':
        df[col] = le.fit_transform(df[col].astype(str))
        print(f"Encoded: {col}")

print(f"\nTotal missing values: {df.isnull().sum().sum()}")
print(f"Data types remaining:")
print(df.dtypes.value_counts())
```

```python
# Convert everything to numeric forcefully
for col in df.columns:
    try:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    except:
        pass

# Fill any NaN created during conversion
df = df.fillna(df.median())

print("Data types after fix:")
print(df.dtypes.value_counts())
print(f"Total missing values: {df.isnull().sum().sum()}")
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Fix all str columns
for col in df.columns:
    if df[col].dtype.name == 'str' or str(df[col].dtype) == 'str':
        df[col] = le.fit_transform(df[col].astype(str))

# Verify
print("Data types after encoding:")
print(df.dtypes.value_counts())
print(f"\nTotal missing values: {df.isnull().sum().sum()}")
```

```python
from sklearn.preprocessing import StandardScaler

# Separate features and target
X = df.drop(columns=['SalePrice', 'Id'])
y = df['SalePrice']

print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print(f"\nFeatures scaled successfully!")
print(f"Mean of scaled features: {X_scaled.mean().mean():.4f} (should be ~0)")
print(f"Std of scaled features: {X_scaled.std().mean():.4f} (should be ~1)")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

print(f"Data split successfully!")
print(f"Training set:   {X_train.shape}")
print(f"Testing set:    {X_test.shape}")
print(f"Training target: {y_train.shape}")
print(f"Testing target:  {y_test.shape}")
```

```python
import joblib

# Save scaled data
X_train.to_csv('../data/X_train.csv', index=False)
X_test.to_csv('../data/X_test.csv', index=False)
y_train.to_csv('../data/y_train.csv', index=False)
y_test.to_csv('../data/y_test.csv', index=False)

# Save scaler for deployment later
joblib.dump(scaler, '../models/scaler.pkl')

print("All data saved!")
print("Scaler saved!")
print("\nFiles saved:")
print("data/X_train.csv")
print("data/X_test.csv")
print("data/y_train.csv")
print("data/y_test.csv")
print("models/scaler.pkl")
```
