```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
pd.set_option('display.max_columns', None)
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

print("All libraries imported successfully!")
```

```python
df = pd.read_csv('../data/train.csv')
print("Dataset Shape:", df.shape)
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

```python
df.head()
```

```python
print("Shape:", df.shape)
print("\nColumn Names:\n", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
```

```python
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
print("Columns with Missing Values:\n")
print(missing)
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['SalePrice'], kde=True, color='blue')
plt.title('Sale Price Distribution')
plt.xlabel('Sale Price')
plt.ylabel('Count')
plt.show()
```

```python
top_corr = df.select_dtypes(include='number').corr()['SalePrice'].sort_values(ascending=False)[1:11]
plt.figure(figsize=(10,6))
sns.barplot(x=top_corr.values, y=top_corr.index, palette='coolwarm')
plt.title('Top 10 Features Correlated with SalePrice')
plt.xlabel('Correlation')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df, color='green', alpha=0.6)
plt.title('Living Area vs Sale Price')
plt.xlabel('Living Area (sqft)')
plt.ylabel('Sale Price')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df, color='green', alpha=0.6)

outliers = df[(df['GrLivArea'] > 4000) & (df['SalePrice'] < 200000)]
plt.scatter(outliers['GrLivArea'], outliers['SalePrice'], 
color='red', s=100, label='Outliers', zorder=5)

plt.title('Living Area vs Sale Price (Outliers Highlighted)')
plt.xlabel('Living Area (sqft)')
plt.ylabel('Sale Price')
plt.legend()
plt.show()

print(f"Number of outliers detected: {len(outliers)}")
print(outliers[['GrLivArea', 'SalePrice']])
```

```python
df = df.drop(df[(df['GrLivArea'] > 4000) & 
                (df['SalePrice'] < 200000)].index)

print(f"Dataset shape after removing outliers: {df.shape}")
print("Outliers removed successfully!")
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='GrLivArea', y='SalePrice', 
                data=df, color='green', alpha=0.6)
plt.title('Living Area vs Sale Price (Outliers Removed)')
plt.xlabel('Living Area (sqft)')
plt.ylabel('Sale Price')
plt.show()

print(f"Final dataset shape: {df.shape}")
```

```python
# Log transform to reduce skewness
import numpy as np

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].set_title('Before Log Transform')
sns.histplot(df['SalePrice'], kde=True, ax=axes[0], color='red')

df['SalePrice'] = np.log1p(df['SalePrice'])

axes[1].set_title('After Log Transform')
sns.histplot(df['SalePrice'], kde=True, ax=axes[1], color='green')

plt.tight_layout()
plt.show()

print("Log Transformation applied successfully!")
```

```python
# Save cleaned data
df.to_csv('../data/train_cleaned.csv', index=False)
print(f"Cleaned data saved!")
print(f"Final shape: {df.shape}")
```
